Importing Important Libraries

```python
import numpy as  np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.set_style('darkgrid')
```

Loading the dataset

```python
df = pd.read_csv(r"C:\Users\91744\Downloads\Finance_data.csv")
```

```python
df.head()
```

Drop the columns that are not required

```python
df.info()
```

```python
df.drop(['Mutual_Funds','Equity_Market','Debentures','Government_Bonds','Fixed_Deposits','PPF','Gold'], axis = 1 , inplace = True)
```

```python
df.head()
```

Replacing the regular expressions

```python
df = df.replace(r'^\s*$', np.nan, regex = True)
```

Now visualizing the data 

```python
sns.countplot(x = 'gender', data = df, linewidth = 3 , palette = "Set2", edgecolor= "black")
plt.show()
```

```python
sns.countplot(x = 'age', data = df , linewidth = 3 , palette ="Set3", edgecolor= "black")
plt.show()
```

```python
plt.figure(figsize = (18,6))
plt.subplot(1,2,1)
sns.countplot(x = df['gender'], hue= df['Investment_Avenues'], palette = 'summer', linewidth = 3, edgecolor = "white")
plt.title("Investment Avenues")

plt.subplot(1,2,2)
sns.countplot(x = df['gender'],hue= df['Stock_Marktet'], palette = 'hot', linewidth = 3 , edgecolor = 'white')
plt.title("Stock Market")
plt.show()
```

```python
sns.countplot(x = df['Factor'], palette = 'coolwarm', linewidth = 3, edgecolor = "black")
plt.show()
```

```python
plt.figure(figsize = (16,8))
sns.pointplot( x = 'Purpose', y = 'age', data = df, linestyles = "--", capsize = .3, color ="blue")
plt.show()
```

```python
plt.figure(figsize = (18,6))
plt.subplot(1,2,1)
sns.countplot( hue= df['gender'], x = df['Duration'], palette = 'coolwarm', linewidth = 2,edgecolor = "black")
plt.title("Duration")

plt.subplot(1,2,2)
sns.countplot( x= df['Invest_Monitor'],palette = 'seismic', linewidth = 2 , edgecolor = "black")
plt.title("Investment Monitoring")
plt.show()
```
